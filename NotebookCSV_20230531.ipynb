```python
#code5-34
#曜日のダミー変数を生成
week_list=['Tue','Wed','Thu','Fri','Sat','Sun','Mon']

enc=OneHotEncoder(categories=[week_list],sparse=False)

week_d=enc.fit_transform(df[['day_of_the_week']])

df_week_d=pd.DataFrame(week_d,index=df.index,columns=week_list,dtype='int')
print(df_week_d)
```

```python
#code5-35
df = pd.concat([df, df_week_d], axis=1)
print(df) #確認
```
