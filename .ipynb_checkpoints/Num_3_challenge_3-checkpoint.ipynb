```python
#001 code3-81
import numpy as np
import pandas as pd
import statsmodels.api as sm

from tbats import TBATS
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error

from pmdarima.model_selection import train_test_split

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize']=[12,9]
plt.rcParams['font.size']=14
```

```python
#002 code3-82
dataset='Num_3_predict/01_output/Num_3_2018_2022_data_1.csv'
df=pd.read_csv(dataset)
df
```

```python
#003 不要な列を削除
df_new=df.drop(df.columns[[0]],axis=1)
df_new
```

```python
#004 code3-83 dataを学習用とテストに分割
train,test=train_test_split(df_new,test_size=258) #1年258回x5年
```

```python
#005 code3-84 データをグラフ
fig, ax=plt.subplots()
ax.plot(train.index,train.values,label='train')
ax.plot(test.index,test.values,label='test',color='gray')
ax.axvline(len(train),color='blue')
ax.legend()
plt.show()
```

```python
#006 code3-85 予想モデルの学習
model=TBATS(seasonal_periods=(258,5)) #1年258回x5年
fitted_model=model.fit(train)
```

```python
#007 code3-86 予測の実施
train_pred=fitted_model.y_hat
test_pred=fitted_model.forecast(steps=258)
```

```python
#008 code3-87 予測モデルのテスト
print('RMSE:\n',np.sqrt(mean_squared_error(test.x,test_pred)))
print('MAE:\n',mean_absolute_error(test.x,test_pred))
print('MAPE:\n',mean_absolute_percentage_error(test,test_pred))
```

```python
#009 code3-88 グラフ化(予測値と実測値)
fig, ax=plt.subplots()
ax.plot(test.index,test.x,label='actual(test)')#実測値
ax.plot(test.index,test_pred,label='predicted')#予測値

ax.legend()
plt.show()
```

```python
#018 code3-97 データの読み込み
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
#019 code3-98 data 読み込み
dataset='Num_3_predict/01_output/Num_3_2018_2022_data_1.csv'
df=pd.read_csv(dataset)
df
```

```python
#020 不要な列を削除
df_new=df.drop(df.columns[[0]],axis=1)
y=df_new
y
```

```python
#021 code3-99 特徴量
lag1=y.shift(1)

window258=lag1.rolling(window=258).mean()

expanding=lag1.expanding().mean()
```

```python
#022 code3-100 特徴量を結合したデータ生成
df_tbl=pd.concat([y,lag1,window258,expanding],axis=1)
df_tbl.columns=['y','lag1','window258','expanding']
print(df_tbl)
```

```python
#023 code3-101 三角関数特徴量
exog=pd.DataFrame()
exog.index=y.index

def fourier_terms_gen(seasonal,terms_num):
    
    for num in range(terms_num):
        num=num+1
        sin_colname='sin'+str(seasonal)+'_'+str(num)
        cos_colname='cos'+str(seasonal)+'_'+str(num)
        
        exog[sin_colname]=np.sin(num*2*np.pi*exog.index/seasonal)
        exog[cos_colname]=np.cos(num*2*np.pi*exog.index/seasonal)
        
fourier_terms_gen(seasonal=5,terms_num=10)
fourier_terms_gen(seasonal=258,terms_num=10)
```

```python
#024 code3-102 データの結合と欠損値削除
df_tbl=pd.concat([df_tbl,exog],axis=1)
df_tbl_1=df_tbl.dropna()
print(df_tbl_1)
```

```python
#025 code3-103 csv保管
df_tbl_1.to_csv('Num_3_predict/01_output/df_tbl_1.csv')
df_tbl_1
```

```python
#026 code3-104 モジュール
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error

from pmdarima.model_selection import train_test_split

import warnings
warnings.simplefilter('ignore')

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize']=[12,9]
plt.rcParams['font.size']=14
```

```python
#027 code3-105 データセット
dataset=('Num_3_predict/01_output/df_tbl_1.csv')
df=pd.read_csv(dataset)
df
```

```python
#028 code3-106 学習データとテストデータに分割
y_train,y_test=train_test_split(df.y,test_size=258)
X_train,X_test=train_test_split(df.drop('y',axis=1),test_size=258)
```

```python
#029 code3-107 予測モデルの学習　学習データ
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#030 code3-108 予測の実施
train_pred=regressor.predict(X_train)
```

```python
#031 code3-109 予測の実施 テストデータ期間
y_train_new=y_train.copy()

for i in range(len(y_test)):
    X_value=X_test.iloc[i:(i+1),:]
    y_value_pred=regressor.predict(X_value)
    y_value_pred=pd.Series(y_value_pred,index=[X_value.index[0]])
    y_train_new=pd.concat([y_train_new,y_value_pred])
    
    lag1_new=y_train_new.iloc[-1]
    window258_new=y_train_new[-258:].mean()
    expanding_new=y_train_new.mean()
    
    X_test.iloc[(i+1):(i+2),0]=lag1_new
    X_test.iloc[(i+1):(i+2),1]=window258_new
    X_test.iloc[(i+1):(i+2),2]=expanding_new
                       
test_pred=y_train_new[-258:]                       
    
```

```python
#032 code3-110 予測モデルのテスト
print('RMSE:\n',np.sqrt(mean_squared_error(y_test,test_pred)))
print('MAE:\n',mean_absolute_error(y_test,test_pred))
print('MAPE:\n',mean_absolute_percentage_error(y_test,test_pred))
```

```python
#033 code3-111 グラフ(予測値と実測値)
fig, ax=plt.subplots()
ax.plot(y_test.index,y_test.values,label='actual(test)')#実測値
ax.plot(y_test.index,test_pred,label='predicted')#予測値

ax.legend()
plt.show()

```
