```python
#Game 三つ目並べ_3  _____F10でクラスごと_____F11で1行ごと_____　キーでステップ_____
from random import randint
#from random import rand
# import random as RandomRecords
#import tictactoe as TicTacToe
# from tictactoe import TicTacToe, B, O, X, rand
#import random
#randint はランダムな整数を返す
import itertools
import os
import csv

#Game Make
#マスの状態　勝敗判定
B=0 #対戦中とマス空き
O=1 #先手の勝ちと先手
X=2 #後手の勝ちと後手
D=3 #引き分け

KEY=' OX'

# #盤面の表示
class TicTacToe: #〇×game
    def __init__(self, player1, player2):
        self.move = {O: player1, X: player2}

    def show(self, board):
        print()
        for i in range(3):
            print(' ' + '|'.join([KEY[board[i * 3 + j]] for j in range(3)]))
            if i < 2:
                print(' -+-+-')
        print()

#勝敗の判定
    def judge(self, board):            
        patterns = [(0, 1, 2), (3, 4, 5), (6, 7, 8),  # horizontal 行 パターン　#_[c1],[c2],[c3]_の列ようです
                    (0, 3, 6), (1, 4, 7), (2, 5, 8),  # vertical 列
                    (0, 4, 8), (2, 4, 6)]             # diagonal 斜め
        for c1, c2, c3 in patterns: #パターン
            if board[c1] and board[c1] == board[c2] == board[c3]:
                
                return board[c1]
        if B not in board:
            return D #引き分け
        return B #対戦中
    
    def play(self): #Game start
        try:
            while True:
                turn, board = O, [B for _ in range(9)] #範囲は9
                while True:
                    self.show(board)
                    if winner := self.judge(board): #ジャッジ
                        if winner == D:
                            print('--- Draw ---') #引き分け
                        else:
                            print(f'*** {KEY[winner]} win!! ***') #勝ち
                        input('\n>>> Press enter to start new game') #再スタート
                        break
                    print(f'turn : {KEY[turn]}') #順番表示
                    move = self.move[turn](board, turn)
                    board[move] = turn
                    turn = X if turn == O else O #後手の勝ち 先手の勝ち

        except KeyboardInterrupt: #キーボード割り込み除く
            pass

#対戦シミュレータ
    def simulate(self, num=100): #まねる
    #def simulate(self): #まねる    
        count, ret, records = 0, [0 for _ in range(4)], []  #記録
        while True:
            turn, board, record = O, [B for _ in range(9)], []
            while True:
                if winner := self.judge(board): #ジャッジ
                    ret[winner] += 1  #1を代入
                    count += 1  #1を代入
                    records += [[winner] + record]
                    break
                move = self.move[turn](board, turn)
                board[move] = turn
                record += [move]
                turn = X if turn == O else O
            if count >= num: #対戦結果
                print(f'O = {ret[O]}/{num} ({ret[O]/num*100:.1f}%)')
                print(f'X = {ret[X]}/{num} ({ret[X]/num*100:.1f}%)')
                print(f'D = {ret[D]}/{num} ({ret[D]/num*100:.1f}%)')
                break
        return records
    

# #user入力_のとき
# def user(board, turn):
#     while True:
#         try:
#             move = int(input('> '))
#             if move < len(board) and not board[move]: #ボードに置かれていないとき
#                 return move
#         except ValueError: #エラーは除外
#             pass

#ランダムなAI
def rand(board, turn): #無作為
    blank = [i for i, cell in enumerate(board) if cell == B] #__boardの空きマスからカウントする__
    rand_index = randint(0, len(blank) - 1) #無作為検索__範囲外のインデックス作成を避けるために -1 を追加する必要があります__
    #blank_rand_index = randint(0, len(blank) - 1)
    
    return blank[rand_index]


# #対戦結果
# #AIと対人戦
# if __name__ == '__main__':
#     tictactoe = TicTacToe(user, rand)
#     tictactoe.play()

#     #対人戦のとき
#     #tictactoe = TicTacToe(user, user)　
#     #tictactoe.play()

# #10回分のランダムなAI対戦結果
# if __name__ == '__main__':
#     tictactoe = TicTacToe(rand, rand)
#     for record in tictactoe.simulate(100):
#         print(record)  
#from tictactoe import TicTacToe, B, O, X, rand

#スコア
O_WIN = 10
X_WIN = -10
DRAW = 0


class RandomRecords:
    def __init__(self):
        self.t = TicTacToe(rand, rand) #ランダムな対戦

    def scoring(self, num): #スコア計算
        scores = {}
        records = self.t.simulate(num) #シュミレートを代入
        for record in records:
            winner = record.pop(0) #勝敗を取得
            score = O_WIN if winner == O else X_WIN if winner == X else DRAW #スコア10,-10,0を取得
            turn, board = O, [B] * 9
            for move in record:
                board[move] = turn
                key = tuple(board)
                if key not in scores:
                    scores[key] = {'total': 0, 'count': 0}
                scores[key]['total'] += score #合計にスコア代入
                scores[key]['count'] += 1 #カウントに1代入
                turn = X if turn == O else O
        return scores


# if __name__ == '__main__':
#     records = RandomRecords()
#     for key, value in records.scoring(1).items():
#         print(key, value)


#パターンの設定
class Patterns:
    def __init__(self, scores):
        self.scores = scores #データを配列化
        self.patterns = {
            'h1': (0, 1, 2),
            'h2': (3, 4, 5),
            'h3': (6, 7, 8),
            'v1': (0, 3, 6),
            'v2': (1, 4, 7),
            'v3': (2, 5, 8),
            'd1': (0, 4, 8),
            'd2': (2, 4, 6),

        }
        self.indexs = {}

    #パターンをインデックスに変換(1～26の)
    def to_index(self, cells):
        repeat = len(cells)
        if repeat not in self.indexs:
            product = enumerate(itertools.product([B, O, X], repeat=repeat))
            self.indexs[repeat] = {j: i for i, j in product}
            #self.indexs[repeat] = {i: j for j, i in product} #出力かわらない

            # df=self.indexs[repeat]
            # print(df)#インデックス確認_追加

        return self.indexs[repeat][cells]

    #集計したデータをファイル形式で出力
    def get_header(self):
        header = ['ave']
        for key, value in self.patterns.items():
            header += [key + '-' + str(i) for i in range(3**len(value))]
        return header

    def get_dataset(self, board, score):
        dataset = {'ave': score['total'] / score['count']}
        for key, value in self.patterns.items():
            pattern = tuple([board[i] for i in value])
            if key not in dataset:
                dataset[key] = [0 for _ in range(3**len(value))]
            dataset[key][self.to_index(pattern)] = 1
        return dataset

    def to_csv(self, name):
        with open(name, 'w', newline='') as f:
            w = csv.writer(f)
            header = self.get_header()
            w.writerow(header)
            scores = sorted(self.scores.items(),
                            key=lambda x: x[1]['total'] / x[1]['count'])
            for board, score in scores:
                dataset = self.get_dataset(board, score)
                line = [dataset['ave']]
                for key in self.patterns.keys():
                    line += dataset[key]
                w.writerow(line)

#AIとAIを対戦[10,100,1000,10000]
if __name__ == '__main__':
    rand = RandomRecords()
    for num in [10, 100, 1000, 10000, 50000, 100000, 150000, 200000, 300000, 500000, 1000000]:
        name = 'rand' + str(num) + '.csv'
        print(f'\n[{name}]')
        scores = rand.scoring(num)
        patterns = Patterns(scores)
        #patterns.to_csv(name)


        data_dir="/workdir/Test_1"
        patterns.to_csv(os.path.join(data_dir,name))
        

        #df=patterns(name) 
        #patterns.to_csv(os.path.join(data_dir,name),index=False)
        #df=patterns.to_csv(name)
        # Num_3_all.to_csv(os.path.join(output_dir,data_file),index=False)

        # output_file=os.path.join(output_dir,'order_data.csv')
        # order_data.to_csv(output_file,index=False)
```

```python
#Game 三つ目並べ_4
from random import randint
#from random import rand
# import random as RandomRecords
#import tictactoe as TicTacToe
# from tictactoe import TicTacToe, B, O, X, rand
#import random
#randint はランダムな整数を返す
import itertools
import os
import csv
import pandas as pd
from sklearn.linear_model import LinearRegression


#Game Make
#マスの状態　勝敗判定
B=0 #対戦中とマス空き
O=1 #先手の勝ちと先手
X=2 #後手の勝ちと後手
D=3 #引き分け

KEY=' OX'

# #盤面の表示
class TicTacToe: #〇×game
    def __init__(self, player1, player2):
        self.move = {O: player1, X: player2}

    def show(self, board):
        print()
        for i in range(3):
            print(' ' + '|'.join([KEY[board[i * 3 + j]] for j in range(3)]))
            if i < 2:
                print(' -+-+-')
        print()

#勝敗の判定
    def judge(self, board):            
        patterns = [(0, 1, 2), (3, 4, 5), (6, 7, 8),  # horizontal 行 パターン　#_[c1],[c2],[c3]_の列ようです
                    (0, 3, 6), (1, 4, 7), (2, 5, 8),  # vertical 列
                    (0, 4, 8), (2, 4, 6)]             # diagonal 斜め
        for c1, c2, c3 in patterns: #パターン
            if board[c1] and board[c1] == board[c2] == board[c3]:
                
                return board[c1]
        if B not in board:
            return D #引き分け
        return B #対戦中
    
    def play(self): #Game start
        try:
            while True:
                turn, board = O, [B for _ in range(9)] #範囲は9
                while True:
                    self.show(board)
                    if winner := self.judge(board): #ジャッジ
                        if winner == D:
                            print('--- Draw ---') #引き分け
                        else:
                            print(f'*** {KEY[winner]} win!! ***') #勝ち
                        input('\n>>> Press enter to start new game') #再スタート
                        break
                    print(f'turn : {KEY[turn]}') #順番表示
                    move = self.move[turn](board, turn)
                    board[move] = turn
                    turn = X if turn == O else O #後手の勝ち 先手の勝ち

        except KeyboardInterrupt: #キーボード割り込み除く
            pass

#対戦シミュレータ
    def simulate(self, num=100): #まねる
        count, ret, records = 0, [0 for _ in range(4)], []  #記録
        while True:
            turn, board, record = O, [B for _ in range(9)], []
            while True:
                if winner := self.judge(board): #ジャッジ
                    ret[winner] += 1  #1を代入
                    count += 1  #1を代入
                    records += [[winner] + record]
                    break
                move = self.move[turn](board, turn)
                board[move] = turn
                record += [move]
                turn = X if turn == O else O
            if count >= num: #対戦結果
                print(f'O = {ret[O]}/{num} ({ret[O]/num*100:.1f}%)')
                print(f'X = {ret[X]}/{num} ({ret[X]/num*100:.1f}%)')
                print(f'D = {ret[D]}/{num} ({ret[D]/num*100:.1f}%)')
                break
        return records
    

# #user入力_のとき
# def user(board, turn):
#     while True:
#         try:
#             move = int(input('> '))
#             if move < len(board) and not board[move]: #ボードに置かれていないとき
#                 return move
#         except ValueError: #エラーは除外
#             pass

#ランダムなAI
def rand(board, turn): #無作為
    blank = [i for i, cell in enumerate(board) if cell == B] #__boardの空きマスからカウントする__
    rand_index = randint(0, len(blank) - 1) #無作為検索__範囲外のインデックス作成を避けるために -1 を追加する必要があります__
    #blank_rand_index = randint(0, len(blank) - 1)
    
    return blank[rand_index]


# #対戦結果
# #AIと対人戦
# if __name__ == '__main__':
#     tictactoe = TicTacToe(user, rand)
#     tictactoe.play()

#     #対人戦のとき
#     #tictactoe = TicTacToe(user, user)　
#     #tictactoe.play()

# #10回分のランダムなAI対戦結果
# if __name__ == '__main__':
#     tictactoe = TicTacToe(rand, rand)
#     for record in tictactoe.simulate(100):
#         print(record)  
#from tictactoe import TicTacToe, B, O, X, rand

#スコア
O_WIN = 10
X_WIN = -10
DRAW = 0


class RandomRecords:
    def __init__(self):
        self.t = TicTacToe(rand, rand) #ランダムな対戦

    def scoring(self, num): #スコア計算
        scores = {}
        records = self.t.simulate(num) #シュミレートを代入
        for record in records:
            winner = record.pop(0) #勝敗を取得
            score = O_WIN if winner == O else X_WIN if winner == X else DRAW #スコア10,-10,0を取得
            turn, board = O, [B] * 9
            for move in record:
                board[move] = turn
                key = tuple(board)
                if key not in scores:
                    scores[key] = {'total': 0, 'count': 0}
                scores[key]['total'] += score #合計にスコア代入
                scores[key]['count'] += 1 #カウントに1代入
                turn = X if turn == O else O
        return scores


# if __name__ == '__main__':
#     records = RandomRecords()
#     for key, value in records.scoring(1).items():
#         print(key, value)



#パターンの設定
class Patterns:
    def __init__(self, scores):
        self.scores = scores #データを配列化
        self.patterns = {
            'h1': (0, 1, 2),
            'h2': (3, 4, 5),
            'h3': (6, 7, 8),
            'v1': (0, 3, 6),
            'v2': (1, 4, 7),
            'v3': (2, 5, 8),
            'd1': (0, 4, 8),
            'd2': (2, 4, 6),
        }
        self.indexs = {}

    #パターンをインデックスに変換(1～26の)
    def to_index(self, cells):
        repeat = len(cells)
        if repeat not in self.indexs:
            product = enumerate(itertools.product([B, O, X], repeat=repeat))
            self.indexs[repeat] = {j: i for i, j in product}
        return self.indexs[repeat][cells]

    #集計したデータをファイル形式で出力
    def get_header(self):
        header = ['ave']
        for key, value in self.patterns.items():
            header += [key + '-' + str(i) for i in range(3**len(value))]
        return header

    def get_dataset(self, board, score):
        dataset = {'ave': score['total'] / score['count']}
        for key, value in self.patterns.items():
            pattern = tuple([board[i] for i in value])
            if key not in dataset:
                dataset[key] = [0 for _ in range(3**len(value))]
            dataset[key][self.to_index(pattern)] = 1
        return dataset

    def to_csv(self, name):
        with open(name, 'w', newline='') as f:
            w = csv.writer(f)
            header = self.get_header()
            w.writerow(header)
            scores = sorted(self.scores.items(),
                            key=lambda x: x[1]['total'] / x[1]['count'])
            for board, score in scores:
                dataset = self.get_dataset(board, score)
                line = [dataset['ave']]
                for key in self.patterns.keys():
                    line += dataset[key]
                w.writerow(line)


# if __name__ == '__main__':
#     rand = RandomRecords()
#     for num in [10, 100, 1000, 10000]:
#         name = 'rand' + str(num) + '.csv'
#         print(f'\n[{name}]')
#         scores = rand.scoring(num)
#         patterns = Patterns(scores)
#         #patterns.to_csv(name)


#         data_dir="/workdir/Test_1"
#         patterns.to_csv(os.path.join(data_dir,name))
        


#回帰分析
class Regression:
    def __init__(self, csv, show=True): #得点表示

        data_dir="/workdir/Test_1"
        dataset = pd.read_csv(os.path.join(data_dir,csv))    

        #dataset = pd.read_csv(csv)
        dataset_except_ave = dataset.drop('ave', axis=1)
        x = dataset_except_ave.values
        y = dataset['ave'].values
        self.model = LinearRegression()
        self.model.fit(x, y)
        self.p = Patterns(None)
        self.show = show

    def _print(self, *args):
        if self.show: #表示しますか？
            print(*args)

    def to_x(self, board): #入力された盤面を目的変数に変換
        x = []
        for key, value in self.p.patterns.items():
            state = tuple([board[i] for i in value])
            patterns = [0 for _ in range(3**len(value))]
            patterns[self.p.to_index(state)] = 1
            x += patterns
        return [x]

    def move(self, board, turn): #回帰式で得点を予測
        blank = [i for i, cell in enumerate(board) if cell == B]
        index, score = None, 0
        self._print()
        for i in blank:
            board[i] = turn
            predict = self.model.predict(self.to_x(board))[0]
            cond = predict > score if turn == O else predict < score
            if index is None or cond:
                index, score = i, predict
            self._print(i, '=', predict)
            board[i] = B
        self._print('move :', index)
        return index


if __name__ == '__main__': #対戦結果__1000回学習させたとき
    num=1000
    # num = 10000 #10000にしてみたけどいうほど勝率UPしなかった

    for name in ['rand10.csv', 'rand100.csv', 'rand1000.csv', 'rand10000.csv','rand50000.csv','rand100000.csv','rand150000.csv','rand200000.csv','rand300000.csv','rand500000.csv','rand1000000.csv']:
        model = Regression(name, show=False)
        print(f'\n[{name}]')
        print('--- O ---')
        TicTacToe(model.move, rand).simulate(num)
        print('--- X ---')
        TicTacToe(rand, model.move).simulate(num)

#200000回の勝率が高いので採用        

```
